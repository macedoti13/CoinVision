```python
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('fotos/img1.png')
img = cv2.resize(img, (720, 720))
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img = cv2.GaussianBlur(img, (3,3), 0)
plt.figure(figsize=(15,15))
plt.imshow(img)
plt.axis('off')
plt.show()
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
plt.figure(figsize=(15,15))
plt.imshow(img_gray, cmap="gray")
plt.axis('off')
plt.show()
```

```python
edges = cv2.Canny(img_gray, 140, 280)
plt.figure(figsize=(15,15))
plt.imshow(edges, cmap='gray')
plt.axis('off')
plt.show()
```

```python
for i in range(1, 11):
    img = cv2.imread(f'fotos/img{i}.png')
    img = cv2.resize(img, (720, 720))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    img = cv2.GaussianBlur(img, (3,3), 0)
    img_gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
    edges = cv2.Canny(img_gray, 200, 400)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    areas = [cv2.contourArea(x) for x in contours]
    card = [x for x in contours if 10000 < cv2.contourArea(x) < 90000]
    card_size = (8.56, 5.398)
    x, y, w, h = cv2.boundingRect(card[0])
    ppcm = h / card_size[0]
    print(ppcm)
    
```
