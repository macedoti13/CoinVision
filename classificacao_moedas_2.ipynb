```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_scaled_radius(scale_factor, original_radius):
    return int(round(original_radius * scale_factor))
```

```python
img = cv2.imread('fotos/img1.png')
img = cv2.resize(img, (1080, 1080))
img = cv2.GaussianBlur(img, (9,9), 0)
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
scale_factor = 59.22897196261682
```

```python
original_radius = 1.1
scaled_radius = calculate_scaled_radius(scale_factor, original_radius)
print(scaled_radius)
```

```python
circles = cv2.HoughCircles(img_gray, cv2.HOUGH_GRADIENT, 1, 30,
                           param1=200, param2=45)

circles
```

```python
if circles is not None:
    circles = np.uint16(np.around(circles))
    for i in circles[0, :]:
        img_rgb = cv2.circle(img_rgb, (i[0], i[1]), i[2], (0, 255, 0), 4)
        
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```
